Найти интервалы, на которых функция возрастает и убывает

```python
if is_increasing(y, Interval(3, oo))==False: print(f'функция возрастает на интервале от 3 до oo') 
if is_increasing(y, Interval(-3, 0))==True: print(f'функция возрастает на интервале от -3 до 0')
if is_decreasing(y, Interval(-oo,-3))==True: print(f'функция убывает интервале от  -oo до -3')
if is_decreasing(y, Interval(0, 3))==True: print(f'функция убывает интервале от 0 до 3')
```

Определить промежутки, на котором f > 0 и на котором f < 0

```python
def intervals(roots):
    intervals=[]
    intervals.append([roots[0]-1, roots[0]])
    for i in range(len(roots)-1):
        j=[roots[i],roots[i+1]]
        intervals.append(j)
    intervals.append([roots[-1], roots[-1]+1]) 
    return intervals
```

```python
interval=intervals(roots)
```

```python
for i in interval:
    start = -oo if interval.index(i)==0 else i[0]
    end = oo if interval.index(i)==len(interval)-1 else i[1]
    if y.evalf(subs={x:(i[0]+i[1])/2}) >0:
        print(f'f > 0 на промежутке {start, end}')
    else:
        print(f'f < 0 на промежутке {start, end}')
```
